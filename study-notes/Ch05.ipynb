### 平面直角坐标系
在平面上，**笛卡尔坐标系**(Cartesian coordinate system)也叫作平面直角坐标系。  
平面直角坐标系是两个相较于**原点**(origin)相互垂直的实数轴。  
数学中，平面直角坐标系场记作$\mathbb{R}^2$。  
**横轴**(horizontal number line)常被称作$x$轴($x$-axis),  
**纵轴**(vertical number line)常被称作$y$轴($y$-axis)。  
※本书中，也以$x_1$表示横轴，$x_2$表示纵轴。

![carteisan_coordinate_system](./images/ch05/cartesian_coordinate_system.png)

横纵轴$xy$将平面($xy$-plane)分成四个**象限**(quadrants)，以**罗马数字**(Roman numeral)**逆时针方向**(counter-clockwise)编号。  
※象限不包括坐标轴。

平面上的点多以坐标($a,b$)表示。$a$和$b$分别是**横坐标**($x$-coordinate)和**纵坐标**($y$-coordinate)。

```python
# Bk3_Ch5_01
import matplotlib.pyplot as plt
import numpy as np

# 定义坐标点
x = [4,  0, -5, -4,  6, 0]
y = [2, -2,  7, -6, -5, 0]

# 定做坐标点的标签
labels = ['A', 'B', 'C', 'D', 'E', 'O']

# 定义绘图用对象
fig, ax = plt.subplots()

# 声明绘制方式。点上的符号为x
# plt.plot(x, y, 'x') # x表示用╳表示一个点
plt.plot(x, y, 'o')  # o表示用●表示一个点

# 用zip()将labels，x，y打包，然后在每次循环里将他们中的元素赋值给label，i，j
for label, i, j in zip(labels, x, y):
    plt.text(i, j+0.5, label+f' ({i},{j})')

# 定义坐标轴的标签为x和y
# plt.xlabel('x轴')
# plt.ylabel('y轴')
plt.xlabel('x')
plt.ylabel('y')

# 定义横纵坐标的起始点及外观。
# y=0，表示从纵轴的0坐标开始画横轴。x=0亦然
plt.axhline(y=0, color='k', linestyle='-')
plt.axvline(x=0, color='k', linestyle='-')
# 定义横纵轴上的刻度
plt.xticks(np.arange(-8, 8+1, step=1))
plt.yticks(np.arange(-8, 8+1, step=1))
# 设置坐标是等比的
plt.axis('scaled')

# 设置横纵轴绘制极限
ax.set_xlim(-8, 8)
ax.set_ylim(-8, 8)
# ax.set_ylim(-10, 10)

# 隐藏图像上下左右的边框
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)

#设置背景网格
ax.grid(linestyle='--', linewidth=0.25, color=[0.5, 0.5, 0.5])
```

### 直线
任意两点可以画一条直线，对应代数中的二元一次方程。
$$
ax+by+c=0
$$
用矩阵乘法，可以写作
$$
\begin{bmatrix}
a&b
\end{bmatrix}
\begin{bmatrix}
x\\
y
\end{bmatrix}
+c=0
$$

平面直角坐标系中的三类直线：  
![straight_line_in_rectangular_plane_coordinate_system](./images/ch05/straight_line_in_rectangular_plane_coordinate_system.png)

如上图(a)所示，当$a=0$时，直线平行于横轴：
$$
by+c=0
$$
如上图(b)所示，当$b=0$时，直线平行于纵轴：
$$
ax+c=0
$$
如上图(c)所示，如果$a、b$均不为0，可以写成：
$$
y=-{a \over b}x-{c \over b}
$$
当$x$为自变量，$y$为因变量时，上式实际上就成了一元一次函数。其中，$-a/b$为直线**斜率**(slope)，$-c/b$为**纵轴截距**($y$-intercept)

### 两点距离
直角坐标系两点，$A(x_A,y_A)$和$B(x_B,y_B)$，之间的距离可以用勾股定理获得：
$$
AB=\sqrt{(x_A-x_B)^2+(y_A-y_B)^2}
$$

### 正圆
以$A(x_A,y_A)$点为圆心，$r$为半径画一个圆。圆上任意一点为$(x,y)$，据此可以构造等式：
$$
\sqrt{(x-x_A)^2+(y-y_A)^2}=r
$$
两边平方得到圆的解析式：
$$
(x-x_A)^2+(y-y_A)^2=r^2
$$
使用矩阵乘法，可写作：
$$
\begin{bmatrix}
x-x_A&y-y_A
\end{bmatrix}
\begin{bmatrix}
x-x_A\\
y-y_A
\end{bmatrix}-r^2=0
$$
特别的，当圆心为原点(0,0)，半径$r$=1是，圆为**单位圆**(unit circle)，对应解析式为：
$$
x^2+y^2=r^2
$$
使用矩阵乘法可以写成：
$$
\begin{bmatrix}
x&y
\end{bmatrix}
\begin{bmatrix}
x\\
y
\end{bmatrix}-r^2=0
$$

### 垂直
两条直线$ax+by+c=0$和$\alpha x+\beta y+\gamma=0$，两者垂直满足如下条件：
$$
a\alpha+b\beta=0
$$
如果系数$a、b、\alpha、\beta$均不为0时，两条直线垂直，则两条直线斜率相乘为-1，即：
$$
{a \over b}{\alpha \over \beta}=-1
$$
如下图所示：  
![two_straight_lines_are_orthogonal](./images/ch05/two_straight_lines_are_orthogonal.png)

### 平行
两条直线$ax+by+c=0$和$\alpha x+\beta y+ \gamma=0$平行，系数满足：
$$
a\beta-b\alpha=0
$$
如果系数$a、b、\alpha、\beta$均不为0，两条直线若平行或重合，则两个斜率相同，即：
$$
{a \over b }={\alpha \over \beta}
$$
如下图所示：  
![two_straight_lines_are_parallel](./images/ch05/two_straight_lines_are_parallel.png)

下图为两条水平线和两条竖直线。两条水平先可以视作常数函数，而两条竖直线则不是函数。  
![two_horizontal_lines_and_two_vertical_lines](./images/ch05/two_horizontal_lines_and_two_vertical_lines.png)

```python
# Bk3_Ch05_2

import numpy as np
import matplotlib.pyplot as plt

# 装饰用函数
def fig_decor(ax):
    # 设置x轴和y轴
    ax.set_xlabel('x')
    ax.set_ylabel('y')

    # 设置水平线位置，范围
    ax.hlines(y=0,xmin=-8,xmax=8,color='k')
    # 设置竖线位置，范围
    ax.vlines(x=0,ymin=-8,ymax=8,color='k')

    # 设置x，y轴刻度，最小，最大，步长
    ax.set_xticks(np.arange(-8,8+1,step=1))
    ax.set_yticks(np.arange(-8,8+1,step=1))

    # 设置x,y轴等比缩放
    ax.axis('scaled')
    # 这是背景方格
    ax.grid(linestyle='--', linewidth=0.25,color=[0.5,0.5,0.5])

    # 设置x，y轴边界
    ax.set_xbound(lower=-8,upper=8)
    ax.set_ybound(lower=-8,upper=8)
    # 隐藏上下左右的边框
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['left'].set_visible(False)

# 用numpy生成画线用的数组
x_array = np.linspace(-8,8)
y_array = np.linspace(-8,8)

# orthogonal
fig, ax=plt.subplots()
plt.title('orthogonal')
# 根据x数组中，根据函数计算对应y的数值
y1=0.5*x_array+2
y2=-2*x_array-1

# 从每个x_array中取出x，从y1，y2中取出y，作为一个点(x,y)，连续画出数组中所有的点，形成直线
ax.plot(x_array,y1)
ax.plot(x_array,y2)
fig_decor(ax)

# parallel
fig, ax=plt.subplots()
plt.title('parallel')
y1=0.5*x_array+2
y2=0.5*x_array-4

ax.plot(x_array,y1)
ax.plot(x_array,y2)
fig_decor(ax)

# horizontal
fig,ax=plt.subplots()
plt.title('horizontal')

y1=0*x_array+2
y2=0*x_array-4

ax.plot(x_array,y1)
# axhline
ax.plot(x_array,y2)
fig_decor(ax)

# vertical
fig,ax=plt.subplots()
plt.title('vertical')
x1=0*y_array+2
x2=0*y_array-4

ax.plot(x1,y_array)
# axvline
ax.plot(x2,y_array)
fig_decor(ax)
```

## 5.3 图解“鸡兔同笼”问题

### 图解法
$x_1$代表鸡，$x_2$代表兔。  
鸡、兔共有35个头，对应等式：
$$
x_1+x_2=35
$$

有94只足，对应等式：
$$
2x_1+4x_2=94
$$

联立两个等式，得到方程组：
$$
\begin{cases}
x_1+x_2=35\\
2x_1+4x_2=94
\end{cases}
$$

如下图，这两条直线的交点，就是解(23,12)。
![image_of_the_system_of_equations_for_the_chicken_and_rabbit_problem](./images/image_of_the_system_of_equations_for_the_chicken_and_rabbit_problem.png)

### 限制条件
$x_1$和$x_2$应均为非负整数。  
故，图像上所有点应被限制在第一象限(包含坐标轴)，这个区域对应不等式：
$$
\begin{cases}
x_1 \ge 0\\
x_2 \ge 0
\end{cases}
$$
如下图所示：  
![non_negative_integer_image_of_the_system_of_equations_for_the_chicken_and_rabbit_problem](./images/ch05/non_negative_integer_image_of_the_system_of_equations_for_the_chicken_and_rabbit_problem.png)

从另一个角度看，上图绿蓝两组点对应的横、纵轴坐标值分别构成**等差数列**(arithmetic progression)

### 二元一次方程组解的个数

两个二元一次方程构成的方程组可以有一个解、无数解或者没有解。  

如下图(a)，两条直线相交于一点，也就是有一个解。  
如下图(b)，两条直线重合，也就是有无数个解。  
如下图(c)，两条直线平行，也就是没有解。  
![number_of_solutions_to_a_system_of_quadratic_equations](./images/ch05/number_of_solutions_to_a_system_of_quadratic_equations.png)

```python
# Bk3_Ch5_03
import numpy as np
import matplotlib.pyplot as plt

# 
plt.rcParams["figure.figsize"]=[3.50, 3.50]
# 紧凑图像边缘
plt.rcParams["figure.autolayout"]=True

x_array=np.arange(0,30+1,step=1)
y_line_1=35-x_array
y_line_2=(94-2*x_array)/4

fig,ax=plt.subplots()

plt.plot(x_array, y_line_1, color='#0070C0')
plt.plot(x_array, y_line_2, color='g')

# solution of linear equations
plt.plot(23,12,marker='x',markersize=12)
plt.axvline(x=23,color='r',linestyle='--')
plt.axhline(y=12,color='r',linestyle='--')

plt.xlabel('$x_1$ (number of chickens)')
plt.ylabel('$x_2$ (number of rabits)')
plt.axhline(y=0,color='k',linestyle='-')
plt.axvline(x=0,color='k',linestyle='-')
plt.xticks(np.arange(0,30+1,step=5))
plt.yticks(np.arange(0,30+1,step=5))
plt.axis('scaled')
# 显示小刻度
plt.minorticks_on()
# 绘制小刻度的的背景线条
ax.grid(which='minor',linestyle=':',linewidth='0.5',color=[0.8,0.8,0.8])
ax.set_xlim(0,30)
ax.set_ylim(0,30)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.grid(linestyle='--',linewidth=0.25,color=[0.5,0.5,0.5])
```

## 5.4 极坐标：距离和夹角

**极坐标系**$\text{(polar coordinate system)}$，也是常用坐标系。如下图所示：  
![polar_coordinate_system_sample](./images/ch05/polar_coordinate_system_sample.png)  
$O$是极坐标的**极点**(pole)，  
从$O$向右引一条射线作为**极轴**$\text{(polar axis)}$，规定逆时针角度为正。  
一般，$r$为**极径**$\text{(radial coordinate}$ 或 $\text{radial distance)}$，  
$\theta$称为**极角**$\text{(angular coordinate}$ 或 $\text{polar angle}$ 或 $\text{azimuth)}$。  
这样，屏幕上任意一点P的位置，可以表示为$(r,\theta)$

平面上，极坐标$(r,\theta)$可以转化为直角坐标系坐标$(x_1,x_2)$：
$$
\begin{cases}
x_1=r\cdot cos\theta \\
x_2=r\cdot sin\theta
\end{cases}
$$

```python
# Bk3_Ch5_04
import numpy as np
import matplotlib.pyplot as plt

def plot_polar(theta, r):
    # set draw type is polar coordinate
    fig,ax=plt.subplots(subplot_kw={'projection':'polar'})
    ax.plot(theta,r)
    # set radial axis limit
    ax.set_rmax(20)
    # set radial axis ticks
    ax.set_rticks([5,10,15,20])
    # position radial labels
    ax.set_rlabel_position(-45)
    ax.set_thetagrids(np.arange(0.0,360.0,60.0))
    plt.show()
  
# %% circle
# 角度从0度到6pi度，生成1000个。生成的越多图像约精细
theta = np.linspace(0,6*np.pi,1000)

r=10+theta*0
plot_polar(theta,r)

# %%Archimedes' spiral
r=1*theta
plot_polar(theta,r)

# %% Rose
# 6是变化的速率，可以表示花瓣个数
r = 10*np.cos(6*theta)+10
plot_polar(theta,r)
```

## 5.5 参数方程：引入一个参数
在平面直角坐标系中，曲线上任意一点$x,y$是某个参数$t$的函数，无论$t$取任何值，方程组确定的点都在曲线上，那么这个方程就叫做**参数方程**$\text{(parametric equation)}$：
$$
\begin{cases}
x_1=f(t)\\
x_2=g(t)
\end{cases}
$$

参数方程表示单位圆：
$$
\begin{cases}
x_1=\text{cos}(t)\\
x_2=\text{sin}(t)
\end{cases}
$$
其中，$t$为参数，取值范围为$[0,2\pi]$

```python
# Bk3_Ch5_05
import numpy as np
import matplotlib.pyplot as plt

t=np.linspace(0,2*np.pi,100)

# parametric equation of unit circle
x1=np.cos(t)
x2=np.sin(t)

fig,ax = plt.subplots()

# plot the cricle
plt.plot(x1,x2)

# plt.show()
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_xticks(np.arange(-2,2+1,step=1))
ax.set_yticks(np.arange(-2,2+1,step=1))
ax.axis('scaled')
ax.grid(linestyle='--',linewidth=0.25,color=[0.7,0.7,0.7])
ax.set_xbound(lower=-2,upper=2)
ax.set_ybound(lower=-2,upper=2)
# plt.gca() => ax
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)
plt.axhline(y=0,color='k',linestyle='-')
plt.axvline(x=0,color='k',linestyle='-')

plt.show()
```
